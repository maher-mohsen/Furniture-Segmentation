# Load And Run

```python
import os
HOME = os.getcwd()
print(HOME)
```

```python
import torch
import cv2
import numpy as np
from ultralytics import YOLO
```

```python
image_path = f'{HOME}/sample.jpg'
output_path = f'{HOME}/output'
print(image_path)
print(output_path)
```

```python
model = YOLO(f'{HOME}/yolov8x-seg.pt')
```

```python
results = model.predict(source=image_path, conf=0.25)
```

```python
results[0].show()
```

```python
import cv2
results[0].save(f"{output_path}/sample.jpg")
print(f"Saved predicted image to: {output_path}")
```
